```python
def generate_confidence_score(row):
    parent_diag_code_flag = row['parent_diag_code_flag']
    exact_diag_code_flag = row['exact_diag_code_flag']
    prior_treatment_flag = row['prior_treatment_flag']
    rx_date_match_flag = row['rx_date_match_flag']
    lag_fill_date_flag = row['lag_fill_date_flag']
    payor_name_final_flag = row['payor_name_final_flag']
    payor_type_final_flag = row['payor_type_final_flag']
    pbm_final_flag = row['pbm_final_flag']

    diagnosis_score_parent = 0
    diagnosis_score_exact = 0

    if parent_diag_code_flag == 1:
        diagnosis_score_parent = 5
    
    if exact_diag_code_flag == 1:
        diagnosis_score_exact = 5

    final_dispense = 0

    if rx_date_match_flag == 1:
        if lag_fill_date_flag == 1 or lag_fill_date_flag == -1:
            final_dispense = 20

    therapy_score = 0

    if prior_treatment_flag == 1:
        therapy_score = 10

    payor_type_score = 0
    payor_name_score = 0
    pbm_score = 0

    if payor_name_final_flag == 1:
        payor_name_score = 2

    if payor_type_final_flag == 1:
        payor_type_score = 5 

    if pbm_final_flag == 1:
        pbm_score = 3

    confidence_score = diagnosis_score_parent + diagnosis_score_exact + final_dispense + therapy_score + payor_type_score + payor_name_score + pbm_score + 50

    return confidence_score
```

```python
df_90['confidence_score'] = df_90.apply(
    generate_confidence_score,
    axis=1
)

df_90.head()
```

```python
df_90["best_match_flag"] = 0

def flag_best(group):
    # Step 1: max confidence rows
    max_conf = group["confidence_score"].max()
    g = group[group["confidence_score"] == max_conf]

    # If max confidence is unique, assign 1 directly
    if len(g) == 1:
        return g.index

    # Step 2: max confidence is tied, use diagnosis_days_lag
    # If all tied rows have inf lag, none should get 1
    if np.isinf(g["diagnosis_days_lag"]).all():
        return pd.Index([])

    # Exclude inf while finding minimum lag
    valid_lag = g[~np.isinf(g["diagnosis_days_lag"])]

    # If somehow no valid lag remains, none gets 1
    if valid_lag.empty:
        return pd.Index([])

    # Minimum lag among tied max-confidence rows
    min_lag = valid_lag["diagnosis_days_lag"].min()

    # All rows with that minimum lag get 1
    return g[g["diagnosis_days_lag"] == min_lag].index


best_indices = (
    df_90.groupby("sp_patient_id", group_keys=False)
         .apply(flag_best)
         .explode()
         .dropna()
         .astype(int)
)

df_90.loc[best_indices, "best_match_flag"] = 1
```
